```python
#AbhinavPal22MIC0024
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Load the dataset
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
#AbhinavPal22MIC0024
# Reshape to include the channel dimension (1 for grayscale)
train_images = train_images.reshape((60000, 28, 28, 1))
test_images = test_images.reshape((10000, 28, 28, 1))

# Normalize the pixel values (scale to 0-1)
train_images = train_images.astype('float32') / 255
test_images = test_images.astype('float32') / 255

# Optional: Convert labels to one-hot encoding (if using 'categorical_crossentropy' loss)
# If using 'sparse_categorical_crossentropy', skip this. The example model below uses sparse.
# from tensorflow.keras.utils import to_categorical
# train_labels = to_categorical(train_labels)
# test_labels = to_categorical(test_labels)
```

```python
model = Sequential([
    # First Convolutional Block
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D((2, 2)),

    # Second Convolutional Block
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),

    # Classification Layers
    Flatten(),
    Dense(100, activation='relu'), # Hidden layer
    Dense(10, activation='softmax')  # Output layer (10 classes for digits 0-9)
])

# Print a summary of the model layers and parameters
model.summary()
```

```python
#AbhinavPal22MIC0024
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(train_images, train_labels, epochs=5, batch_size=64, validation_split=0.1)
```

```python
#AbhinavPal22MIC0024
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=2)
print(f'\nTest accuracy: {test_acc:.4f}')
```
